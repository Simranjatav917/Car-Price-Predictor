```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
#Data collection
car_dataset = pd.read_csv('car data.csv')
```

```python
car_dataset.head()
```

```python
car_dataset.Transmission.value_counts()
```

```python
# encoding "Fuel_Type" Column
car_dataset.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)

# encoding "Seller_Type" Column
car_dataset.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)

# encoding "Transmission" Column
car_dataset.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
car_dataset.head()
```

```python
#Split data into training data and test data
X = car_dataset.drop(['Car_Name','Selling_Price'],axis=1)
Y = car_dataset['Selling_Price']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, random_state=2)
```

```python
#Model Training

##Linear Regression
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

```python
#Prediction
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
#R square error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared error: ",error_score)
```

```python
#Visualize the actual price and predicted prices

plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
# prediction on Training data
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
##LEGO REGRESSION
```

```python
# loading the linear regression model
lass_reg_model = Lasso()
```

```python
lass_reg_model.fit(X_train,Y_train)
```

```python
# prediction on Training data
training_data_prediction = lass_reg_model.predict(X_train)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
# prediction on Training data
test_data_prediction = lass_reg_model.predict(X_test)
```

```python
# R squared Error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
#Frequency of cars vs the Feul type
#0->Petrol
#1->Desiel
#2->CNG

#Frequency of cars vs transmission type
#0-> Manual
#1->Automatic


plt.subplot(1,4,2)
plt1 = car_dataset.Fuel_Type.value_counts().plot(kind='bar')
plt.title('Fuel Type Histogram')
plt1.set(xlabel = 'Fuel Type', ylabel='Frequency of fuel type')

plt.subplot(1,4,4)
plt1 = car_dataset.Transmission.value_counts().plot(kind='bar')
plt.title('Car Transmission Histogram')
plt1.set(xlabel = 'Transmission', ylabel='Frequency of Car type')

plt.show()
```

```python
#Average price vs engine transmission type
#0-> Manual
#1->Automatic

#Average Price vs Fuel Type
#0->Petrol
#1->Desiel
#2->CNG


df = pd.DataFrame(car_dataset.groupby(['Transmission'])['Selling_Price'].mean().sort_values(ascending = False))
df.plot.bar()
plt.title('Transmission Type vs Average Price')
plt.show()

df = pd.DataFrame(car_dataset.groupby(['Fuel_Type'])['Selling_Price'].mean().sort_values(ascending = False))
df.plot.bar()
plt.title('Fuel Type vs Average Price')
plt.show()
```
